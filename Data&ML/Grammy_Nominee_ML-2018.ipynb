```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import os
import numpy as np
```

```python
df = pd.read_csv(os.path.join("Data API/Nomination_ML_url_2010_2017.CSV"))
df.columns
```

```python
#CLEAN DATA
df = df.drop(['features.analysis_url', 'features.id','features.track_href',
              'features.type', 'features.uri','duration_ms','song_id',
              'track','song_name','artist','release_date','features.time_signature','features.mode','artist_image_url_l','artist_image_url_s','album_url','artist_popularity'], axis=1)
df.head()
```

```python
target = df['win/loss']
target_names = ["loss", "win"]
```

```python
data = df.drop('win/loss', axis=1)
feature_names = data.columns
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators=300)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
print("test score: ",rf.score(X_test, y_test))
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
df_2018 = pd.read_csv(os.path.join("Data API/2018_top_song_features_images_url.CSV"))
df_2018_1 =df_2018.drop(['features.analysis_url', 'features.id','features.track_href','features.type', 'features.uri','duration_ms','song_id','track','song_name','artist','release_date','features.time_signature','features.mode','artist_image_url_l','artist_image_url_s','album_url','artist_popularity'], axis=1)
df_2018_1.head(10)
```

```python
#predict with model
y_pred = rf.predict(df_2018_1)
y_pred
```

```python
y_pred = pd.Series(y_pred)
y_pred_proba = rf.predict_proba(df_2018_1)[:,1]
y_pred_proba = pd.Series(y_pred_proba)
max(y_pred_proba)
```

```python
df_2018['Predict_Prob'] = y_pred_proba.values
df_2018['Predict Result']=y_pred.values
```

```python
df_2018 = df_2018.sort_values("Predict_Prob",ascending =False)
df_2018.head(20)
```

```python
#add a Ranking in the DF
df_2018['Rank'] = df_2018['Predict_Prob'].rank(ascending=False)
df_2018['Rank'] = df_2018['Rank'].astype(int)
df_2018.head()
```

```python
#Convert probability to %
df_2018['Predict_Prob'] = pd.Series(["{0:.2f}%".format(val * 100) for val in df_2018['Predict_Prob']], index = df_2018.index)
df_2018
```

```python
df_2018["features.instrumentalness"]=df_2018['features.instrumentalness'].round(4)
df_2018.head()
                                                                                
```

```python
df_2018.to_csv("Data API/2018_Nominee_ML_Predict.CSV",index = False)
```
